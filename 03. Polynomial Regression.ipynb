# 3. Polynomial Regression

### 공부 시간에 따른 시험 점수 (우등생)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data/PolynomialRegressionData.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### 3-1. 단순 선형 회귀 (Simple Linear Regression)

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

### 데이터 시각화 (전체)

```python
plt.scatter(X, y, color='blue')
plt.plot(X, reg.predict(X), color='green')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show
```

```python
reg.score(X, y)
```

### 3-2. 다항 회귀 (Polynomial Regression)

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
X_poly[:5] # [X] -> [X^0, X^1, X^2] -> x 가 3이면 [1, 3, 9]으로 변환
```

```python
X[:5]
```

```python
poly_reg.get_feature_names_out()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

### 데이터 시각화 (변환된 X와 y)

```python
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(poly_reg.fit_transform(X)), color='green')
plt.title('Score by hours(genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show
```

```python
X_range = np.arange(min(X), max(X), 0.1) #X의 최소값에서 최대값까지의 범위를 0.1단위로 잘라서 데이터를 생성
X_range
```

```python
X_range.shape
```

```python
X_range = X_range.reshape(-1, 1) # row 개수는 자동으로 계산하고, column 개수는 1개
X_range.shape
```

```python
X_range[:5]
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X_range, lin_reg.predict(poly_reg.fit_transform(X_range)), color='green')
plt.title('Score by hours(genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show
```

### 공부 시간에 따른 시험 성적 예측

```python
reg.predict([[2]]) # 2시간 공부했을 때 선형 회귀 모델의 예측
```

```python
lin_reg.predict(poly_reg.fit_transform([[2]])) # 2시간을 공부했을 때 다항 회귀 모델의 예측
```

### 과소적합 또는 과대적합에 주의해서 모델을 만들어야 한다.

```python
lin_reg.score(X_poly, y)
```
